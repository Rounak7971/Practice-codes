```python
class jupyter:
    def execute(self):
        print('Compiling')
        print('Running')
class pycharm:
    def execute(self):
        print('Line record')
        print('Debugging')
        print('Compiling')
        print('running')

class program:
    def code(self,ide):
        ide.execute()
py=pycharm()
j=jupyter()
p=program()
p.code(py)
```

```python
class student:
    def __init__(self,m1,m2):
        self.m1=m1
        self.m2=m2
    def __add__(self,other):
        m1=self.m1+other.m1
        m2=self.m2+other.m2
        s3=student(m1,m2)
        return s3
    def __ge__(self,other):
        if self.m1+self.m2>=other.m1+other.m2:
            return True
        else:
            return False
    def __str__(self):
        return '{},{}'.format(self.m1,self.m2)
s1=student(2,88)
s2=student(2,88)

s3=s1+s2
print(s3.m2)

if s1>=s2:
    print( 'S1 wins')
else:
    print('S2 wins')
print(s1)
```

```python
from abc import ABC,abstractmethod
class C(ABC):
    @abstractmethod
    def info(self):
        pass
class Z(C):
    def info(self):
        print('Yeah,it works')

    
    
c=Z()
c.info()
```

```python
class computer(ABC):
    @abstractmethod
    def write(self):
        pass
class whiteboard(computer):
    def write(self):
        print('Writing in progress')
class programmer():
    def code(self,other):
        print('Running code')
        other.write()
wh=whiteboard()
prog=programmer()
prog.code(wh)

    
    
    
```

```python
class rep:
    def __init__(self):
        self.num=1
    def __iter__(self):
        return self
    def __next__(self):
        if self.num<=10:
            val=self.num
            self.num+=1
            return val
        else:
            raise StopIteration
values=rep()

for i in values:
    print(i)

```

```python
def topten():
    yield 5
    yield 6
    yield 7
c=topten()
for i in c:
     print(i)
```

```python
def sqre():
    i=1
    while i<11:
        yield i*i
        i+=1
c=sqre()
for i in c:
    print(i)
```

```python
from threading import *
from time import sleep

class Hello(Thread):
    def run(self):
        for i in range(5):
            print('Hello')
            sleep(1)
class Hi(Thread):
    def run(self):
        for i in range(5):
            print('Hi')
            sleep(1)
o1=Hello()
o2=Hi()
o1.start()
sleep(0.2)
o2.start()
o1.join()
o2.join()
print("bye")
```
